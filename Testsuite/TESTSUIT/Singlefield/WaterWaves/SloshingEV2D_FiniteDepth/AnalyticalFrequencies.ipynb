```python
import numpy as np
from numpy import pi, tanh, sqrt
l = 150 # length in m
h = 30 # depth in m
g = 9.81 # graviatational acceleration
N = 20 # number of modes  
```

# natural frequencies

The wave numbers are
$$ k = \frac{n\pi}{l} $$
The corresponding natural frequencies follow from the dispersion relation
$$ \omega^2 = g k \tanh(kh) $$
which , for deept water ($\lambda < 2g $) simplifies to
$$ \omega^2 = g k $$.

```python
n = np.arange(1,N+1)
k = n*pi/l

omega_exact = sqrt(g*k*tanh(k*h))
omega_deep = sqrt(g*k)

for i in range(len(n)):
    print('%2i: exact = %8.6f Hz, deep = %8.6f Hz'%(n[i],omega_exact[i]/2/pi,omega_deep[i]/2/pi))
    
for i in range(10):
    print('%2i: %8.6f Hz'%(n[i],omega_exact[i]/2/pi))
```
