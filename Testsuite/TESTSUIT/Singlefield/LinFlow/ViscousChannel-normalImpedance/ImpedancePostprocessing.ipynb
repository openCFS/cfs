```python
import h5py
from sys import path
path.append('/home/hhassanp/Devel/CFS_SRC/CFS/share/python/')
from hdf5_tools import get_centroids, get_result, get_coordinates, get_step_values

Pt = get_result('results_hdf5/ViscousChannel-normalImpedance.cfs','fluidMechPressure',step='all',region='S_flow')
Vt = get_result('results_hdf5/ViscousChannel-normalImpedance.cfs','fluidMechVelocity',step='all',region='S_flow')
Xt = get_coordinates('results_hdf5/ViscousChannel-normalImpedance.cfs',region='S_flow')
##line in length of channel
Itop = np.argwhere(np.abs(Xt[:,1]-0.005) < 2e-4).ravel()
Is = np.argsort(Xt[Itop,0])
Itop = Itop[Is]

Pw = get_result('results_hdf5/ViscousChannel-wallEnd.cfs','fluidMechPressure',step='all',region='S_flow')
Vw = get_result('results_hdf5/ViscousChannel-wallEnd.cfs','fluidMechVelocity',step='all',region='S_flow')
Xw = get_coordinates('results_hdf5/ViscousChannel-wallEnd.cfs',region='S_flow')
##line in length of channel
Itopw = np.argwhere(np.abs(Xw[:,1]-0.005) < 2e-4).ravel()
Isw = np.argsort(Xw[Itopw,0])
Itopw = Itopw[Isw]
```

```python
# Pressure Plot

fig,ax = plt.subplots()
ax.plot(Xt[Itop,0],abs(Pt[Itop]),label='Imp-sound hard')
ax.plot(Xw[Itopw,0],abs(Pw[Itopw].T),'--',label='Wall')
ax.set_ylabel('Pressure-abs in Pa')
ax.set_xlabel('Channel distance x in m')
ax.legend(fancybox=True, framealpha=0.5,loc='best')
plt.tight_layout()


# Pressure Plot
fig,ax = plt.subplots()
ax.plot(Xt[Itop,0],Pt[Itop].T.real,label='Imp-sound hard')
ax.plot(Xw[Itop,0],Pw[Itop].T.real,'--',label='Wall')
ax.set_ylabel('Pressure-real in Pa')
ax.set_xlabel('Channel distance x in m')
ax.legend(fancybox=True, framealpha=0.5,loc='best')

# Pressure Plot
fig,ax = plt.subplots()
ax.plot(Xt[Itop,0],Pt[Itop].T.imag,label='Imp-sound hard')
ax.plot(Xw[Itop,0],Pw[Itop].T.imag,'--',label='Wall')
ax.set_ylabel('Pressure-imag in pa')
ax.set_xlabel('Channel distance x in m')

ax.legend(fancybox=True, framealpha=0.5,loc='best')
```

```python
#velocity Plot
## real part of velocity is zero
fig,ax = plt.subplots()
ax.plot(Xt[Itop,0],abs(Vt[Itop,0].T),label='Imp-sound hard')
ax.plot(Xw[Itopw,0],abs(Vw[Itopw,0].T),'--',label='Wall')
ax.set_ylabel('Velocity-abs in m/s')
ax.set_xlabel('Channel distance x in m')
ax.legend(fancybox=True, framealpha=0.5,loc='best')

#velocity Plot
## real part of velocity is zero
fig,ax = plt.subplots()
ax.plot(Xt[Itop,0],Vt[Itop,0].T.imag,label='Imp-sound hard')
ax.plot(Xw[Itopw,0],Vw[Itopw,0].T.imag,'--',label='Wall')
ax.set_ylabel('Velocity-imag in m/s')
ax.set_xlabel('Channel distance x in m')
ax.legend(fancybox=True, framealpha=0.5,loc='best')

#velocity Plot
## real part of velocity is zero
fig,ax = plt.subplots()
ax.plot(Xt[Itop,0],Vt[Itop,0].T.real,label='Imp-sound hard')
ax.plot(Xw[Itopw,0],Vw[Itopw,0].T.real,'--',label='Wall')
ax.set_ylabel('Velocity-real in m/s')
ax.set_xlabel('Channel distance x in m')
#ax.legend()
ax.legend(fancybox=True, framealpha=0.5,loc='best')
```
