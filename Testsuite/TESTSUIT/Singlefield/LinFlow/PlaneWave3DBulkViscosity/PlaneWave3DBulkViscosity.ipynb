```python
# reference file from 2D case
h5r = '../PlaneWave2DBulkViscosity/PlaneWave2DBulkViscosity.h5ref'
Xr = get_coordinates(h5r)
Vr = get_result(h5r,'fluidMechVelocity')
Pr = get_result(h5r,'fluidMechPressure')
Ir = np.argwhere(np.abs(Xr[:,1])<1e-5).ravel()
Ir = Ir[np.argsort(Xr[Ir,0])]
```

```python
h5 = 'results_hdf5/test.cfs'

X = get_coordinates(h5)
V = get_result(h5,'fluidMechVelocity')
P = get_result(h5,'fluidMechPressure')

k = [1,1/10,1/3] # wave vector
k = k/np.linalg.norm(k)

# distance from k-line though 0
Ik = np.argwhere(np.linalg.norm(X-np.outer(np.sum(X*k,axis=1),k),axis=1)<1e-6).ravel()
d = np.linalg.norm(X,axis=1)
Is = Ik[np.argsort(d[Ik])]
Vk = np.sum(V*k,axis=1)

fig,ax = plt.subplots()
l, = ax.plot(d[Is],np.abs(Vk[Is]))
ax.plot(d[Is],Vk[Is].real,'--',color=l.get_color())
ax.plot(d[Is],Vk[Is].imag,':',color=l.get_color())

l, = ax.plot(Xr[Ir,0],np.abs(Vr[Ir,0]),lw=0.3)
ax.plot(Xr[Ir,0],Vr[Ir,0].real,'--',color=l.get_color(),lw=l.get_lw())
ax.plot(Xr[Ir,0],Vr[Ir,0].imag,':',color=l.get_color(),lw=l.get_lw())

fig,ax = plt.subplots()
l, = ax.plot(d[Is],np.abs(P[Is]))
ax.plot(d[Is],P[Is].real,'--',color=l.get_color())
ax.plot(d[Is],P[Is].imag,':',color=l.get_color())

l, = ax.plot(Xr[Ir,0],np.abs(Pr[Ir]),lw=0.3)
ax.plot(Xr[Ir,0],Pr[Ir].real,'--',color=l.get_color(),lw=l.get_lw())
ax.plot(Xr[Ir,0],Pr[Ir].imag,':',color=l.get_color(),lw=l.get_lw())
```
