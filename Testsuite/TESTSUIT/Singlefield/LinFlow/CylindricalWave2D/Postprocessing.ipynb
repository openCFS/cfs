```python
# load the solution
%matplotlib inline
import h5py
from sys import path
#path.append('/home/ftoth/cfs/CFS/share/python/') # add path as needed
path.append('/home/hhassanp/Devel/CFS_SRC/CFS/share/python')
from hdf5_tools import get_centroids, get_result, get_coordinates, get_step_values

Xt1 = get_coordinates('results_hdf5/CylindricalWave2D.cfs','S_1')
Pt1 = get_result('results_hdf5/CylindricalWave2D.cfs','fluidMechPressure',step='all')
Vt1 = get_result('results_hdf5/CylindricalWave2D.cfs','fluidMechVelocity',step='all')

Itopy1 = np.argwhere(np.abs(Xt1[:,1]-0) < 1e-4).ravel()
Isy1 = np.argsort(Xt1[Itopy1,0])
Itopy1 = Itopy1[Isy1]

Itopx1 = np.argwhere(np.abs(Xt1[:,0]-0) < 1e-4).ravel()
Isx1 = np.argsort(Xt1[Itopx1,1])
Itopx1 = Itopx1[Isx1]

Itop1 = np.argwhere(np.abs(Xt1[:,1]-Xt1[:,0])<1e-3).ravel()
Is1 = np.argsort(Xt1[Itop1,0])
Itop1 = Itop1[Is1]

L1 = np.shape(Itop1)
r1 = []
v1 = []
for i in range(L1[0]):
    r1.append(L1)
    v1.append(L1)
    r1[i] = np.sqrt(np.power(Xt1[Itop1[i],0],2) + np.power(Xt1[Itop1[i],1],2))
    v1[i] = np.sqrt(np.power(Vt1[Itop1[i],0],2) + np.power(Vt1[Itop1[i],1],2))
    
fig,ax = plt.subplots()
ax.plot(r1[:],Pt1[Itop1].T.real,'b:',label='At diagonal')
ax.plot(Xt1[Itopy1,0],Pt1[Itopy1].T.real,'c',label='At x axis')
ax.plot(Xt1[Itopx1,1],Pt1[Itopx1].T.real,'r:',label='At y axis')
ax.set_ylabel('Pressure in pa')
ax.set_xlabel('Distance in m')
ax.legend()


fig,ax = plt.subplots()
ax.plot(r1[:],np.abs(v1[:]),'b:',label='At diagonal') 
ax.plot(Xt1[Itopy1,0],abs(Vt1[Itopy1,0]).T.real,'c',label='At x axis')
ax.plot(Xt1[Itopx1,1],abs(Vt1[Itopx1,1]).T.real,'r:',label='At y axis')
ax.set_ylabel('Velocity in m/s')
ax.legend()
```
