```python
import h5py
from sys import path
path.append('/home/hhassanp/Devel/CFS_SRC/CFS/share/python/')
path.append('/home/ftoth/cfs/CFS/share/python/')
from hdf5_tools import get_centroids, get_result, get_coordinates, get_step_values

Pt = get_result('results_hdf5/AcousticChannel.cfs','fluidMechPressure',step='all')
Vt = get_result('results_hdf5/AcousticChannel.cfs','fluidMechVelocity',step='all')
Xt = get_coordinates('results_hdf5/AcousticChannel.cfs')
##line in length of channel
Itop = np.argwhere(np.abs(Xt[:,1]-0.000052) < 2e-6).ravel()
Is = np.argsort(Xt[Itop,0])
Itop = Itop[Is]

##line in width of channel

Itop1 = np.argwhere(np.abs(Xt[:,0]-0.25) < 1e-6).ravel()
Is1 = np.argsort(Xt[Itop1,1])
Itop1 = Itop1[Is1]
```

```python
# load the solution for rotated channel with angel "ang"
ang = np.tan(pi/6)

Xtn = get_coordinates('results_hdf5/PlaneWave2DBulkViscosity-rotated.cfs')
Ptn = get_result('results_hdf5/PlaneWave2DBulkViscosity-rotated.cfs','fluidMechPressure',step='all')
Vtn = get_result('results_hdf5/PlaneWave2DBulkViscosity-rotated.cfs','fluidMechVelocity',step='all')

##line in length of channel
Itopn = np.argwhere(np.abs(Xtn[:,1]-ang*Xtn[:,0]) < 2e-1).ravel()
Isn = np.argsort(Xtn[Itopn,0])
Itopn = Itopn[Isn]

#plot at x = y (diagonal)
L = np.shape(Itopn)
rn = []
vn = []
for i in range(L[0]):
    rn.append(L)
    vn.append(L)    
    rn[i] = np.sqrt(np.power(Xtn[Itopn[i],0],2) + np.power(Xtn[Itopn[i],1],2))
    vn[i] = np.sqrt(np.power(Vtn[Itopn[i],0],2) + np.power(Vtn[Itopn[i],1],2))
```

```python

# Pressure Plot
fig,ax = plt.subplots()
ax.plot(rn[:],abs(Ptn[Itopn].T),'r',label='Rotated-abs')
l,=ax.plot(Xt[Itop,0],abs(Pt[Itop].T),'c--',label='Horizontal-abs',lw=1)

ax.set_ylabel('Pressure in pa')
ax.set_xlabel('Channel distance x in m')
ax.legend()

# Pressure Plot
fig,ax = plt.subplots()
ax.plot(rn[:],Ptn[Itopn].T.real,'r',label='Rotated-Real')
ax.plot(Xt[Itop,0],Pt[Itop].T.real,'--',label='Horizontal-Real',color=l.get_color(),lw=1)

ax.set_ylabel('Pressure in pa')
ax.set_xlabel('Channel distance x in m')
ax.legend()

# Pressure Plot
fig,ax = plt.subplots()
ax.plot(rn[:],Ptn[Itopn].T.imag,'r',label='Rotated-imag')
ax.plot(Xt[Itop,0],Pt[Itop].T.imag,'--',label='Horizontal-imag',color=l.get_color(),lw=1)

ax.set_ylabel('Pressure in pa')
ax.set_xlabel('Channel distance x in m')
ax.legend()
```

```python
#velocity Plot
## real part of velocity is zero
fig,ax = plt.subplots()
ax.plot(rn[:],np.abs(vn),'r',label='Rotated-abs')
l,=ax.plot(Xt[Itop,0],abs(Vt[Itop,0].T),'--',label='Horizontal-abs',lw=1)
ax.set_ylabel('Velocity in m/s')
ax.set_xlabel('Channel distance x in m')
ax.legend()
```
