```python
import matplotlib as mpl
from matplotlib import pyplot as plt
from cycler import cycler
color_order = ['#377eb8','#4daf4a','#e41a1c','#984ea3','#ff7f00','#a65628','#f781bf','#999999','#ffff33']
color_cycle = cycler('color',color_order)
mpl.rc('figure',figsize=[2.7,1.6],dpi=300)
mpl.rc('font',size=8.0)
#mpl.rcParams['font.sans-serif'] = 'Helvetica'
mpl.rc('lines',linewidth=0.7) 
mpl.rc('axes',prop_cycle=color_cycle,grid=True,linewidth=0.5,axisbelow=True,unicode_minus=False)
mpl.rc('grid',linewidth=0.3,linestyle=':')
mpl.rc('legend',fontsize='medium',framealpha=1.0,numpoints=1)
mpl.rc('svg',fonttype='none')
mpl.rc('savefig',dpi=600)
import numpy as np
from numpy import pi
import sys
sys.path.append('../../../../../share/python/') # correct relative path
from hdf5_tools import get_result, get_centroids, get_step_values
```

```python
def readRegions(fun,regions):
    """read multiple regions and combine the results"""
    res = []
    for region in regions:
        res.append(fun(region))
    return np.vstack(res)

h5 = 'results_hdf5/aniso-49.cfs'
regions = ['V_%i'%i for i in [1,2,3]]
C = readRegions( lambda r: get_centroids(h5,region=r), regions )

B = readRegions( lambda r: get_result(h5,'magFluxDensity',region=r,step=1,multistep=1), regions )
J = readRegions( lambda r: get_result(h5,'magTotalCurrentDensity',region=r,step=1,multistep=1), regions )
h5 = 'results_hdf5/iso-1.cfs'
B1 = readRegions( lambda r: get_result(h5,'magFluxDensity',region=r,step=1,multistep=1), regions )
J1 = readRegions( lambda r: get_result(h5,'magTotalCurrentDensity',region=r,step=1,multistep=1), regions )
h5 = 'results_hdf5/iso-4.cfs'
B4 = readRegions( lambda r: get_result(h5,'magFluxDensity',region=r,step=1,multistep=1), regions )
J4 = readRegions( lambda r: get_result(h5,'magTotalCurrentDensity',region=r,step=1,multistep=1), regions )
h5 = 'results_hdf5/iso-9.cfs'
B9 = readRegions( lambda r: get_result(h5,'magFluxDensity',region=r,step=1,multistep=1), regions )
J9 = readRegions( lambda r: get_result(h5,'magTotalCurrentDensity',region=r,step=1,multistep=1), regions )


r = np.sqrt(C[:,0]**2+C[:,1])
phi = np.arctan2(C[:,1],C[:,0])
Ix = np.argwhere(np.abs(phi-phi.min())<1e-6).ravel()
Ix = Ix[np.argsort(r[Ix])]
Iy = np.argwhere(np.abs(phi-phi.max())<1e-6).ravel()
Iy = Iy[np.argsort(r[Iy])]

fig,ax = plt.subplots()
xval,yval = C[Ix,0],B[Ix,2]
l, = ax.plot(xval, np.abs(yval),lw=0.7 )
ax.plot(xval, yval.real,'--',lw=l.get_lw(),color=l.get_color() )
ax.plot(xval, yval.imag,':',lw=l.get_lw(),color=l.get_color() )

xval,yval = C[Iy,1],B[Iy,2]
l, = ax.plot(xval, np.abs(yval),lw=0.7 )
ax.plot(xval, yval.real,'k--',lw=l.get_lw(),color=l.get_color() )
ax.plot(xval, yval.imag,'k:',lw=l.get_lw(),color=l.get_color() )

#xval,yval = C[Ix,0],B1[Ix,2]
#l, = ax.plot(xval, np.abs(yval),lw=0.5,label='iso-1' )
#ax.plot(xval, yval.real,'--',lw=l.get_lw(),color=l.get_color() )
#ax.plot(xval, yval.imag,':',lw=l.get_lw(),color=l.get_color() )

xval,yval = C[Ix,0],B4[Ix,2]
l, = ax.plot(xval, np.abs(yval),lw=0.5,label='iso-4' )
ax.plot(xval, yval.real,'--',lw=l.get_lw(),color=l.get_color() )
ax.plot(xval, yval.imag,':',lw=l.get_lw(),color=l.get_color() )

xval,yval = C[Ix,0],B9[Ix,2]
l, = ax.plot(xval, np.abs(yval),lw=0.5,label='iso-9' )
ax.plot(xval, yval.real,'--',lw=l.get_lw(),color=l.get_color() )
ax.plot(xval, yval.imag,':',lw=l.get_lw(),color=l.get_color() )

ax.legend()
ax.set_xlabel('Radial coordinate in m')
ax.set_ylabel('Magnetic flux density in T')

# plot the total current density
fig,ax = plt.subplots()

xval,yval = C[Ix,0],J[Ix,1]
l, = ax.plot(xval, np.abs(yval),lw=0.7 )
ax.plot(xval, yval.real,'k--',lw=l.get_lw(),color=l.get_color() )
ax.plot(xval, yval.imag,'k:',lw=l.get_lw(),color=l.get_color() )

xval,yval = C[Iy,1],J[Iy,0]
l, = ax.plot(xval, np.abs(yval),lw=0.7 )
ax.plot(xval, yval.real,'k--',lw=l.get_lw(),color=l.get_color() )
ax.plot(xval, yval.imag,'k:',lw=l.get_lw(),color=l.get_color() )

#xval,yval = C[Ix,0],J1[Ix,1]
#l, = ax.plot(xval, np.abs(yval),lw=0.3,label='iso-1' )
#ax.plot(xval, yval.real,'--',lw=l.get_lw(),color=l.get_color() )
#ax.plot(xval, yval.imag,':',lw=l.get_lw(),color=l.get_color() )

xval,yval = C[Ix,0],J4[Ix,1]
l, = ax.plot(xval, np.abs(yval),lw=0.3,label='iso-4' )
ax.plot(xval, yval.real,'--',lw=l.get_lw(),color=l.get_color() )
ax.plot(xval, yval.imag,':',lw=l.get_lw(),color=l.get_color() )

xval,yval = C[Ix,0],J9[Ix,1]
l, = ax.plot(xval, np.abs(yval),lw=0.3,label='iso-9' )
ax.plot(xval, yval.real,'--',lw=l.get_lw(),color=l.get_color() )
ax.plot(xval, yval.imag,':',lw=l.get_lw(),color=l.get_color() )

ax.legend()
#ax.plot(xval, np.angle(yval),'k-.',lw=l.get_lw(),color=l.get_color() )
#ax.plot(C[:,0],C[:,1],'.')
#ax.plot(C[Ix,0],C[Ix,1])
#ax.plot(C[Iy,0],C[Iy,1])
ax.set_xlabel('Radial coordinate in m')
ax.set_ylabel('Current density in A/m2')
```
