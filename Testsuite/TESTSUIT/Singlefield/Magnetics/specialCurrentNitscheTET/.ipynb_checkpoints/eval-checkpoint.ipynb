```python
import sys
import os
import numpy as np
from numpy import pi
from glob import glob
import subprocess
import xml.etree.ElementTree as ET
import pickle


sys.path.append('/home/kroppert/Devel/CFS_SRC/latest_trunkGit/CFS/share/python/')
```

```python
def evalme(meshSizeUp, meshSizeLow, journalFile, xmlFile):
    subprocess.run(["rm","-r","history","results_hdf5","{0}.info.xml".format(xmlFile)])
    sUp=("mSizeUp="+str(meshSizeUp))
    sLow=("mSizeLow="+str(meshSizeLow))
    print(sUp,sLow)
    subprocess.run(["/share/programs/bin/trelis","-batch","-nographics","-nojournal",sLow,sUp,"{0}.jou".format(journalFile)])
    subprocess.run(["/home/kroppert/Devel/CFS_BIN/latest_trunk_opt/bin/cfs", "-t2","{0}".format(xmlFile)])

    fileUp = 'history/{0}-magEnergy-region-V_up.hist'.format(xmlFile)
    fileLow = 'history/{0}-magEnergy-region-V_low.hist'.format(xmlFile)
    with open (fileUp, 'rt') as in_file:
        magEnergy_up = (float(in_file.readlines()[3].split(' ')[2]))
    with open (fileLow, 'rt') as in_file:
        magEnergy_low = (float(in_file.readlines()[3].split(' ')[2]))
    
    # read number of dof's from info.xml file
    tree = ET.parse('{0}.info.xml'.format(xmlFile))
    root = tree.getroot()
    for elem in tree.iterfind('sequenceStep/OLAS/magneticEdge/setup/feFunctions'):
        dof = (float(elem.attrib['totalNumEqns']))
        
    nc=0
    v=0
    for elem in tree.iterfind('header/domain/regions/region'):
        if( elem.attrib['name'] == 'NC_low'):
            nclow=elem.attrib['elems']
            nc = nc + int(nclow)
        if( elem.attrib['name'] == 'NC_up'):
            ncup=elem.attrib['elems']
            nc = nc + int(ncup)
        if( elem.attrib['name'] == 'V_low'):
            vlow=elem.attrib['elems']
            v = v + int(vlow)
        if( elem.attrib['name'] == 'V_up'):
            vup=elem.attrib['elems']
            v = v + int(vup)
        
    
    totalEnergy = magEnergy_up + magEnergy_low
    
    return magEnergy_up, magEnergy_low, totalEnergy, dof, nc,v
```

# HEX-HEX Conform Mesh: h-convergence for reference solution

Here I over-refine the conform mesh, in order to get a correct reference solution.

The energy for the fines (''converged'') mesh is then my solution for the computation relative error

```python
#meshSize = [0.011, 0.012, 0.015, 0.02, 0.03, 0.05, 0.07, 0.09, 0.11, 0.15, 0.20, 0.25, 0.3]
meshSize = [ 0.02, 0.03, 0.05,0.07, 0.09, 0.11, 0.15, 0.20, 0.25, 0.3, 0.4, 0.5]
meshSize = [ 0.2]

polyOrder=0
intOrder=1

# perform conform h refinement and read the magnetic Energy of bottom and top part
magEnergy_hhc_up=[]
magEnergy_hhc_low=[]
magEnergyTotal_hhc = []
num_nc_hhc=[]
num_v_hhc=[]
dof_hhc=[]
for m in meshSize:
    [up, low, total, dof, num_nc, num_v] = evalme(m, m, 'meshConform_HEX_HEX', 'conform')
    magEnergy_hhc_up.append(up)
    magEnergy_hhc_low.append(low)
    magEnergyTotal_hhc.append(total)
    num_nc_hhc.append(num_nc)
    num_v_hhc.append(num_v)
    dof_hhc.append(dof)
    
    
print('magEnergy up = ',magEnergy_hhc_up)
print('magEnergy low = ',magEnergy_hhc_low)
print('magEnergy up + low = ', magEnergyTotal_hhc)
print('number of nc elems = ',num_nc_hhc)
print('number of c elems = ',num_v_hhc)
print('dofs = ',dof_hhc)



# Save the data
with open('hexhexconform_{0}_{1}.pkl'.format(polyOrder,intOrder), 'wb') as f:
    pickle.dump([magEnergy_hhc_up, magEnergy_hhc_low, magEnergyTotal_hhc, dof_hhc, num_nc_hhc, num_v_hhc], f)
```
