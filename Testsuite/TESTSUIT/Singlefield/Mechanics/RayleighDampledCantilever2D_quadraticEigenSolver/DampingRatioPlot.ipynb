```python
eigsRe = [-853.672, -783.935, -582.696, -578.719, -383.281, -361.02, -254.997, -192.093, -151.219, -83.7712, -78.8753, -43.7142, -23.6599, -22.1019, -16.2296, -15.0346, -15.0346, -16.2296, -22.1019, -23.6599, -43.7142, -78.8753, -83.7712, -151.219, -192.093, -254.997, -361.02, -383.281, -578.719, -582.696, -783.935, -853.672]
eigsIm = [-9118.03, -8733.78, -7512, -7485.79, -6056.5, -5871.25, -4892.31, -4203.85, -3687.69, -2621.09, -2526.13, -1693.96, -930.284, -842.44, -350.287, -56.8461, 56.8461, 350.287, 842.44, 930.284, 1693.96, 2526.13, 2621.09, 3687.69, 4203.85, 4892.31, 5871.25, 6056.5, 7485.79, 7512, 8733.78, 9118.03]
s = np.array([ a+1j*b for a,b in zip(eigsRe,eigsIm) ])
```

```python
sR = s.real/s.imag
sXi = np.sqrt(sR**2/(1+sR**2))
sW0 = s.imag/np.sqrt(1-sXi**2)
sXi
```

```python
alpha = 30
beta = 0.00002

w =2*np.pi*np.linspace(10,2000,200)
xi = alpha/(2*w)+beta*w/2

fig,ax = plt.subplots()
ax.plot(w,xi)
ax.plot(s.imag,sXi,'x')
ax.set_xlim(left=0)
ax.plot(sW0,sXi,'.')
ax.set_xlabel('Frequency')
ax.set_ylabel('Damping ratio')
```
