$\sigma = 5.67\cdot10^{-8}\,\frac{W}{m²K^4}$ ... stefan-boltzmann constant

$\epsilon = 0.32$ ... emissivity

$r = 0.5\,m$ ... radius of the wire

$\dot{q} = 100\,\frac{W}{m³}$ ... volumetric heat source density

$T_{amb} = 293.15\,K$ ... ambient temperature

total heat source: $\dot{Q} = \dot{q}V = \dot{q}AL = \dot{q}\pi r² L \rightarrow \frac{\dot{Q}}{L} = \dot{q}\pi r² = 78.54\,\frac{W}{m}$

radiant power: $\dot{Q} = \epsilon\sigma A\left(T⁴-T_{amb}⁴\right) = \epsilon\sigma\left(T⁴-T_{amb}⁴\right) 2\pi r L \rightarrow \frac{\dot{Q}}{L} = \epsilon\sigma\left(T⁴-T_{amb}⁴\right) 2\pi r$

static state solution: $\dot{q}\pi r² = \epsilon\sigma\left(T⁴-T_{amb}^4\right) 2\pi r \rightarrow T = \sqrt[4]{\frac{\dot{q}r}{2\epsilon\sigma} + T_{amb}⁴} = \sqrt[4]{\frac{100\,\frac{W}{m³}\cdot 0.5\,m}{2\cdot0.32\cdot5.67\cdot10^{-8}\,\frac{W}{m²K⁴}} + (293.15\,K)⁴} = 305.959158851\,K$ on $\Gamma_{\text{thermalradiation}}$
