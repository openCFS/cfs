Due to transmission conditions, there will be some reflection of wave propagation at the acoustic -thermoviscous interfaces . to have a good evaluation of the reflection coefficients and its effects on our results, we conclude to study the value of this reflection coefficient.
In the case of plane wave(e.g. wave propagation in channel), the wave propagates in following form (no attenuation)
\begin{equation}
p(x,t) = p^+ e^{iwt-ikx}+ p^- e^{iwt+ikx}
\end{equation}
where $p^+$ and $p^-$ are the amplitudes of the forward wave incident and the amplitude of back coming(reflected) waves, respectively. The real wave number $k = \omega/c_0$ .

In this sense, reflection coefficient at the interface placing at $x_0$, defines from following equation
\begin{equation}
R = \frac{p^- e^{ikx_0}}{p^+e^{-ikx_0}}.
\label{eq:amp-ref-coeff}
\end{equation}

```python
f = 10e3
pR = 0
pI = 0
K = 1.4261e5
rho = 1.225

c = np.sqrt(K/rho)
k = 2*np.pi*f/c

from scipy.optimize import curve_fit   #we could import more, but this is what we need
###defining your fitfunction
def myFun(x,Ar,Ai,Br,Bi,L):
    A = Ar + 1j * Ai
    B = Br + 1j * Bi
    return A*np.exp((-1j*k-L)*x) + B*np.exp((1j*k+L)*x)

## In order to fit both real and imaginary parts pressure 
## we define two functions for each parts
def funcBoth(x,Ar,Ai,Br,Bi,L):
    p_real = np.real(myFun(x,Ar,Ai,Br,Bi,L))
    p_imag = np.imag(myFun(x,Ar,Ai,Br,Bi,L))
    return np.hstack([p_real, p_imag])

##let us guess some start values
initialGuess=[1,0,1,0,0.01]
##making the actual fit
pBoth = np.hstack([P1[Itop1].real, P1[Itop1].imag])
poptBoth,_ = curve_fit(funcBoth,X1[Itop1,0],pBoth,initialGuess)

#one may want to
fig,ax = plt.subplots()
ax.plot(X1[Itop1,0],P1[Itop1].real,label='Numerical-real')
ax.plot(X1[Itop1,0],P1[Itop1].imag,label='Numerical-imag')
ax.plot(X1[Itop1,0],myFun(X1[Itop1,0],*poptBoth).real,'.',label='Fit-real')
ax.plot(X1[Itop1,0],myFun(X1[Itop1,0],*poptBoth).imag,'.',label='Fit-imag')
ax.legend()
ax.set_xlabel('Length in m')
ax.set_ylabel('Pressure in pa')

pneg = poptBoth[2] +1j* poptBoth[3]
pplus = poptBoth[0] +1j* poptBoth[1] 
## Reflection coef
xinterface = 0.15
lambd = poptBoth[4]
r = (pneg/pplus)*np.exp(2*(1j*k*xinterface+lambd*xinterface))
print("Reflection coef =",r)
print(abs(r))
poptBoth
pneg/pplus
```
