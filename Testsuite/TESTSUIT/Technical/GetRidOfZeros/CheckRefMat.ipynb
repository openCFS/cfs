```python
from scipy.sparse import coo_matrix
from scipy.io import mmread
from matplotlib import pyplot as plt
import numpy as np
```

```python
ref_auto_filename = "auto.ref.mtx"
ref_yes_filename = "yes.ref.mtx"
ref_no_filename = "no.ref.mtx"

ref_auto = mmread(ref_auto_filename)
ref_yes = mmread(ref_yes_filename)
ref_no = mmread(ref_no_filename)

print(str(ref_no.nnz) + " no")
print(str(ref_auto.nnz) + " auto with default tolerance")
print(str(ref_yes.nnz) + " yes with higher tollerance")

plt.spy(ref_no, marker=".", color="b", markersize=7)

plt.spy(ref_auto, marker=".", color="r", markersize=5)

plt.spy(ref_yes, marker=".", color="g", markersize=3)
plt.show()
```

```python
default_tolerance = 1.0e-20
xml_tolerance = 1e-14

data = ref_no.data
row = ref_no.row
col = ref_no.col

mask_yes = np.abs(data) > xml_tolerance

filtered_row_yes = row[mask_yes]
filtered_col_yes = col[mask_yes]
filtered_data_yes = data[mask_yes]

filtered_yes = coo_matrix((filtered_data_yes, (filtered_row_yes, filtered_col_yes)), shape=ref_no.shape)

if((filtered_yes!=ref_yes).nnz == 0):
    print("\"yes\" reference matrix is correct (if \"no\" was correct)")
else:
    print("SOMETHING IS WRONG WITH \"yes\"")

# as solution for "auto" case is static and not harmonic,
# we can just check if there are any enteties below tolerance and compare static and harminic cases approximatly

if(not (np.any(np.abs(ref_auto.data) < default_tolerance))):
    print("for \"auto\" no values below the default tolerance are detected")
else:
    print("SOMETHING IS WRONG WITH \"auto\"")


mask_auto = np.abs(data) > default_tolerance
###
filtered_row = row[mask_auto]
filtered_col = col[mask_auto]
filtered_data = data[mask_auto]

filtered_auto = coo_matrix((filtered_data, (filtered_row, filtered_col)), shape=ref_no.shape)

if((filtered_auto!=ref_auto).nnz == 0):
    print("\"auto\" reference matrix is correct (if \"no\" was correct)")
else:
    print("CHECK CAREFULLY as the \"auto\" matrices does not match. There are " + str((filtered_auto!=ref_auto).nnz) + " not matching elements")
    print("It is not a problem as the system matrices are generated for different solvers. One of them is complex and another is real only")
    
```
